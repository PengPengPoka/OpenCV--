```python
import cv2 as cv  # opencv lib
import os

from matplotlib import pyplot as plt
```

```python
directory = os.getcwd()                         # get path to directory
image_data = directory + "\\utils\\Lenna.png"   # path to image

directory
image_data
```

```python
image = cv.imread(image_data)
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
b, g, r = cv.split(image)
```

```python
blue_histogram = cv.calcHist([b], [0], None, [256],[0,256])     # calculate blue channel histogram
green_histogram = cv.calcHist([g], [0], None, [256],[0,256])    # calculate green channel histogram
red_histogram = cv.calcHist([r], [0], None, [256],[0,256])      # calculate red channel histogram
```

```python
plt.plot(blue_histogram, color='blue')
plt.plot(green_histogram, color='green')
plt.plot(red_histogram, color='red')
plt.xlabel("Pixel intensity")
plt.ylabel("Frequency")
plt.title("Intensity of image")
plt.legend(['blue', 'green', 'red'])
plt.show()
```

```python
grayscale = cv.cvtColor(image,cv.COLOR_BGR2GRAY)

gray_histogram = cv.calcHist([grayscale], [0], None, [256], [0,256])
```

```python
plt.plot(gray_histogram)
plt.xlabel("Pixel intensity")
plt.ylabel("Frequency")
plt.title("Intensity of image")

plt.show()
plt.imshow(grayscale)
```

```python
ret, threshold = cv.threshold(grayscale, 120, 255, cv.THRESH_BINARY)
plt.imshow(threshold)
```
